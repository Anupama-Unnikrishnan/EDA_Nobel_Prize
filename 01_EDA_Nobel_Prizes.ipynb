## 3.DATA CLEANING

```python
#TO FILL MISSING VALUES IN 'sex' COLUMN
df['sex'].fillna('Unknown', inplace=True)
```

```python
#TO FILL MISSING VALUES IN 'birth_date' COLUMN
df['birth_date'].fillna('Unknown', inplace=True)
```

```python
#TO FILL MISSING VALUES IN 'birth_city' AND 'birth_country'
df['birth_city'].fillna('Unknown', inplace=True)
df['birth_country'].fillna('Unknown', inplace=True)
```

```python
#TO FILL MISSING VALUES IN 'death_date', 'death_city', 'death_country'
df['death_date'].fillna('Alive', inplace=True)
df['death_city'].fillna('Unknown', inplace=True)
df['death_country'].fillna('Unknown', inplace=True)
```

```python
#TO FILL MISSING VALUES IN AFFILIATION COLUMNS
df['affiliation_name'].fillna('Unknown', inplace=True)
df['affiliation_city'].fillna('Unknown', inplace=True)
df['affiliation_country'].fillna('Unknown', inplace=True)
```

```python
#TO FILL MISSING VALUES IN WIKIPEDIA URL
df['wikipedia_url'].fillna('Not Available', inplace=True)
```

```python
#TO FILL MISSING VALUES IN COUNTRY WINNER RANK
df['country_winner_rank'].fillna(0, inplace=True)
```

```python
#TO VERIFY THAT MISSING VALUES ARE FILLED
print(df.isnull().sum())
```

```python
#TO REMOVE DUPLICATE ROWS IF ANY
df = df.drop_duplicates()

#TO CHECK REMAINING DUPLICATES
print("Number of duplicate rows after cleaning:", df.duplicated().sum())
```

### DATA CLEANING SUMMARY

- Removed duplicate rows from the dataset.
- Filled missing values in important columns such as gender, birth details, death details, affiliation information, and Wikipedia URLs.
- Replaced missing numeric values in `country_winner_rank` with 0.

## 7.FEATURE ENGINEERING

Feature engineering is the process of creating new meaningful features from existing data to improve analysis and reveal hidden patterns. It helps simplify complex information, enhance visualizations, and prepare data for machine learning. In this project, features such as decade, is_female, is_alive, shared_prize, and has_wikipedia were created to support deeper trend and pattern analysis.


### CREATE A 'DECADE' COLUMN

```python
#TO CREATE A NEW COLUMN 'DECADE'
df['decade'] = (df['award_year'] // 10) * 10

print(df[['award_year', 'decade']].head())
```

### CREATE A NEW COLUMN 'IS_FEMALE'

```python
#TO CREATE A NEW COLUMN TO IDENTIFY FEMALE WINNERS
df['is_female'] = df['sex'].apply(lambda x: 1 if x.lower() == 'female' else 0)

print(df[['sex', 'is_female']].head())
```

### CREATE A NEW COLUMN 'IS_ALIVE'

```python
#TO CREATE A COLUMN THAT INDICATES WHETHER THE LAUREATE IS STILL ALIVE
df['is_alive'] = df['death_date'].apply(lambda x: 0 if x != 'Alive' else 1)

print(df[['death_date', 'is_alive']].head())
```

### CREATE A COLUMN FOR SHARED PRIZES

```python
#TO CREATE A COLUMN FOR SHARED PRIZE IDENTIFICATION
df['shared_prize'] = df['is_shared'].apply(lambda x: 'Shared' if x == 1 else 'Not Shared')

print(df[['is_shared', 'shared_prize']].head())
```

### CREATE A COLUMN ‘HAS_WIKIPEDIA’

```python
#TO CHECK WHETHER A WIKIPEDIA PAGE EXISTS
df['has_wikipedia'] = df['wikipedia_url'].apply(lambda x: 0 if x == 'Not Available' else 1)

print(df[['wikipedia_url', 'has_wikipedia']].head())
```

## Feature Engineering Summary

### Feature Engineering Summary
1. **Decade** → Groups award years into 10-year intervals for trend analysis.  
2. **Is_Female** → Converts gender into numeric form (1 = Female, 0 = Others).  
3. **Is_Alive** → Identifies whether the laureate is alive or deceased.  
4. **Shared_Prize** → Indicates whether the prize was shared between multiple laureates.  
5. **Has_Wikipedia** → Shows visibility of laureates on the internet.  

## 8.OUTLIER DETECTION

Outlier detection is the process of identifying unusual data points that differ greatly from the majority of values in a dataset. Outliers can occur due to rare events, natural variability, errors, or special circumstances. Detecting outliers helps understand data distribution, identify anomalies, and improve the accuracy of analysis. In the Nobel Prize dataset, outliers mainly appear in prize amounts and shared prizes, but they are kept because they represent real historical variations.


### OUTLIER DETECTION USING BOXPLOT

```python
#TO CHECK OUTLIERS IN PRIZE AMOUNT USING BOXPLOT
plt.figure(figsize=(10,5))
sns.boxplot(x=df['prize_amount'],color='green')
plt.title("OUTLIERS IN PRIZE AMOUNT")
plt.xlabel("PRIZE AMOUNT")
plt.show()
```

### OUTLIER DETECTION FOR WINNERS PER YEAR

```python
#OUTLIERS IN NUMBER OF WINNERS PER YEAR
plt.figure(figsize=(10,5))
sns.boxplot(x=df['winners_per_year'],color='lightgreen')
plt.title("OUTLIERS IN WINNERS PER YEAR")
plt.xlabel("WINNERS PER YEAR")
plt.show()
```

### OUTLIER DETECTION USING IQR METHOD (Mathematical)

```python
#IQR METHOD FOR PRIZE AMOUNT
Q1 = df['prize_amount'].quantile(0.25)
Q3 = df['prize_amount'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)

#IDENTIFY OUTLIERS
outliers = df[(df['prize_amount'] < lower_limit) | (df['prize_amount'] > upper_limit)]
print("Number of Outliers:", len(outliers))
outliers.head()
```

### OUTLIERS IN PRIZE AMOUNT ADJUSTED

```python
#BOXPLOT FOR ADJUSTED PRIZE AMOUNT
plt.figure(figsize=(10,5))
sns.boxplot(x=df['prize_amount_adjusted'],color='salmon')
plt.title("OUTLIERS IN ADJUSTED PRIZE AMOUNT")
plt.xlabel("PRIZE AMOUNT ADJUSTED (INFLATION)")
plt.show()
```

### OUTLIER ANALYSIS SUMMARY

1. **Prize Amount Outliers:**  
   - Many outliers due to historical changes in prize amounts.  
   - These outliers are valid and should NOT be removed.

2. **Winners Per Year Outliers:**  
   - Some years have multiple shared winners.  
   - These are legitimate variations.

3. **Prize Amount Adjusted:**  
   - Inflation causes large differences → naturally many outliers.

### Decision:
- Outliers are meaningful in the context of Nobel Prize data.  
- Therefore, **we keep all outliers** and DO NOT remove them.
